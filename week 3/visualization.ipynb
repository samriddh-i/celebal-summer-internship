```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("50_Startups (1).csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Profit'], kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot R&D Spend vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='R&D Spend', y='Profit', data=data)
plt.title('R&D Spend vs Profit')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Plot Administration Spend vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Administration', y='Profit', data=data)
plt.title('Administration Spend vs Profit')
plt.xlabel('Administration Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Plot Marketing Spend vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing Spend', y='Profit', data=data)
plt.title('Marketing Spend vs Profit')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Plot Profit by State
plt.figure(figsize=(10, 6))
sns.boxplot(x='State', y='Profit', data=data)
plt.title('Profit by State')
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```
